```python
# create a column with state population for each county
statepop = temp.groupby(['Year', 'State'])['PopulationTotal'].sum().reset_index(name='Pop')
display(statepop)
```

```python
temp = pd.merge(temp, statepop)
display(temp)
```

```python
# estimating how many lottery tickets have been bought per person
# lottery average price = $2 
# (lottery sales / 2) / statePopulation * population score = lottery tickets sold in county

def estimateLottery(x, y, z):
  # x = sales in dollars
  # y = pop score
  # z = state pop
  x = x/2
  x = x/z
  r = x*y
  return r

result['Lottery Tickets Sold'] = temp.apply(lambda x: estimateLottery(x['Lottery Sales $'], x['PopulationTotal'], x['Pop']), axis=1)
display(result.head(25))
```

```python
# winning odds estimation
# 100 million = 1/302575350
# 1 million = 1/12607306
# 10'000 = 1/931001
# 500 = 1/38792
# 200 = 1/14547
# 10 = 640
# 4 = 1/89
# 2 = 1/37

o = [302575350, 12607306, 931001, 38792, 14547, 640, 89, 37]
w = [100000000, 1000000, 10000, 500, 200, 10, 4, 2]

# ticketsSold / odds
# sum(choose random tickets sold out of odds)

import random

def randomWinner(tickets, odds):
  if tickets > odds:
    return int(tickets/odds)
  else:
    # simulate lottery
    a = (random.randint(1, int(odds/tickets)))
    b = (random.randint(1, int(odds/tickets)))
    if a==b:
      return 1
    else:
      return 0

def estimateWinnings(tickets):
  jackpot = 100000000
  million = 1000000
  tenk = 10000
  fiveh = 500
  twoh = 200
  ten = 10
  four = 4
  two = 2
  total = 0

  # estimating up to bottom
  for e in range(len(o)):
    r = randomWinner(tickets, o[e])
    total = total+(r * w[e])

  return total

def adjustWin(winnings, tax):
  if tax != 0:
    winnings = winnings - (winnings * tax/100)
  return winnings

result['Lottery Winnings'] = result['Lottery Tickets Sold'].apply(estimateWinnings)
result['Lottery Winnings'] = result.apply(lambda x: adjustWin(x['Lottery Winnings'], x['Win Tax %']), axis=1)
display(result.head(25))
```
