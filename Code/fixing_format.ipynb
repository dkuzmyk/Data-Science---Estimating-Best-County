```python
# adjust the format of the dataset so all datasets are the same
import pandas as pd
import numpy as np

df = pd.read_csv('Unemployment_by_county.csv')


df = df.drop(['Unnamed: 0'], axis=1)


df.head(5)
```

```python
df2010 = df[['State', 'County', '2010']]
df2011 = df[['State', 'County', '2011']]
df2012 = df[['State', 'County', '2012']]
df2013 = df[['State', 'County', '2013']]
df2014 = df[['State', 'County', '2014']]
df2015 = df[['State', 'County', '2015']]
df2016 = df[['State', 'County', '2016']]
df2017 = df[['State', 'County', '2017']]
df2018 = df[['State', 'County', '2018']]
df2019 = df[['State', 'County', '2019']]

datasets = [df2010, df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019, ]
```

```python
for e in range(len(datasets)):
  datasets[e]['Year'] = datasets[e].columns[2]
  datasets[e].columns = ['State', 'County', 'Unemployment', 'Year']

datasets[4].head(5)
```

```python
r = pd.concat(datasets, ignore_index=True)
r.to_csv('Unemployment.csv')
```
