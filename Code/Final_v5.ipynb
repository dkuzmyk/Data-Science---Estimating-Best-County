```python
file_name = "FINAL_V4.csv"

all_data = pd.read_csv(file_name)

#display(load_data)

load_data = all_data[['State', 'County', 'Year', 'Lottery Sales $', 'PopulationScore']]

display(load_data)
```

```python
the_years = load_data['Year'].unique().tolist()
the_states = load_data['State'].unique().tolist()
the_data = []
# rename
merged = load_data

# enumerate by year
for j, i in enumerate(the_years):
    # group by year
    grouped = merged[merged['Year'] == i]
    
    avg = float(grouped['Lottery Sales $'].mean())
    
    for m,k in enumerate(the_states):
        # get value for state
        the_state_amt = grouped['Lottery Sales $']
        
        # get the sales sum for that specific state at that specific year
        grouped_by_states_in_the_specific_year = pd.DataFrame(grouped[grouped['State'] == k])
        
        #display(grouped_by_states_in_the_specific_year)
        
        tmp = grouped_by_states_in_the_specific_year
        
        tmp['SalesScore'] = the_state_amt/avg * tmp['PopulationScore']
        tmp['SalesScore'] = tmp['SalesScore'] * (m / int(len(tmp))+1)
        
        # save data into array
        the_data.append(tmp)
        
display(the_data[0])
```

```python
final_data = pd.DataFrame([])
for i in the_data:
    final_data = pd.concat([final_data, i], axis = 0)
    
finish_data = all_data.merge(final_data, how='inner')

#file_name = "FINAL_v5.csv"

finish_data = finish_data[['State', 'County', 'Year', 'SalesScore']]

file_name = "save_tmp.csv"

finish_data.to_csv(file_name, index=False)
```

# Select the right columns to use

```python
selection = the_dataset_to_use[['State', 'County', 'Year', 
                                'Median_Income', 'PopulationScore',
                                'Tax', 'Win Tax %',
                                'CrimeScore', 'SalesScore']]

selection['StatesTaxScore'] = -1 * round(selection['Tax'] * selection['Median_Income'] / 100 /100, 2)

selection['WinTaxScore'] = 10 - selection['Win Tax %'] / 10

selection['StatesTaxScore'] = -1*(selection['Median_Income']/ 100 / 100)*selection['WinTaxScore']+selection['StatesTaxScore']

other_data = []

# enumerate by year
for j, i in enumerate(the_years):
    # group by year
    grouped = selection[selection['Year'] == i]
    
    avg = float(selection['StatesTaxScore'].mean())
    
    for m,k in enumerate(the_states):
        # get value for state
        the_state_amt = grouped['StatesTaxScore']
        
        # get the sales sum for that specific state at that specific year
        grouped_by_states_in_the_specific_year = pd.DataFrame(grouped[grouped['State'] == k])
        
        # apply the formulas
        
        # shorten the name of the variable
        tmp = grouped_by_states_in_the_specific_year
        
        # update the taxscore with normalized values
        tmp['StatesTaxScore'] = ((the_state_amt)/float(avg+0.0001)) * -1
        
        # apply the total score formula
        tmp['TotalScore'] = tmp['CrimeScore'] + tmp['StatesTaxScore'] + tmp['SalesScore']
        
        # save data into array holder
        other_data.append(tmp)
        
display(other_data[0])
```

```python
final_data = pd.DataFrame([])
for i in other_data:
    final_data = pd.concat([final_data, i], axis = 0)
    
finish_data = the_dataset_to_use.merge(final_data, how='inner')

finish_data.to_csv('FINAL_v7.csv', index=False)
```
