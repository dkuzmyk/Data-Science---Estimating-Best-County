```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('crime 2010-2019.xls')
```

```python
df = df.drop(columns=['Combined Violent and Property Crime Offenses Known to Law Enforcement by County (Known Offenses)', 'Unnamed: 2'])
df.columns = ['n', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
display(df)
```

```python
# replace nan with 0
df = df.fillna(0)
# drop first row
df = df.drop(df.index[0])
display(df)
```

```python
# split county into county and state
df[['County','State']] = df.n.str.split(", ",expand=True)
df = df.drop(columns=['n'])

df = df[['State', 'County', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']]

display(df)
```

```python
df2010 = df[['State', 'County', '2010']] 
df2011 = df[['State', 'County', '2011']] 
df2012 = df[['State', 'County', '2012']] 
df2013 = df[['State', 'County', '2013']] 
df2014 = df[['State', 'County', '2014']] 
df2015 = df[['State', 'County', '2015']] 
df2016 = df[['State', 'County', '2016']] 
df2017 = df[['State', 'County', '2017']] 
df2018 = df[['State', 'County', '2018']] 
df2019 = df[['State', 'County', '2019']] 

s = [df2010, df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019]
```

```python
display(s[0])
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]

for e in range(len(s)):
  s[e].columns = ['State', 'County', 'TotalCrime']
  s[e]['Year'] = years[e]
```

```python
for e in range(len(s)):
  if e == 0:
    result = s[e] 
  else:
    result = pd.concat([result, s[e]], ignore_index=True)

display(result)
```

```python
result.to_csv('crime.csv', index=False)
```

```python
crimedf = pd.read_csv('crime.csv')
final = pd.read_csv('FINAL_v3.csv')
```

```python
r = pd.merge(final, result)
display(r)
```

```python
# calculate crime score

def crimeScore(countyCrime, stateCrime, countyPop):
  if countyCrime != 0:
    r = countyCrime/stateCrime
    r = r*(-1)*countyPop
  else:
    return countyCrime

  return r
```

```python
display(r)
```

```python
# estimate state crime
stateCrime = r.groupby(['State', 'Year'])['TotalCrime'].sum().reset_index(name='StateCrime')
r = pd.merge(r, stateCrime)
display(r)
```

```python
test = r
```

```python
test['CrimeScore'] = r.apply(lambda x: crimeScore(x['TotalCrime'], x['StateCrime'], x['PopulationScore']), axis=1)
display(test)
```

```python
test.to_csv('FINAL_v4.csv', index=False)
```
